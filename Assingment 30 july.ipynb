## 9 . average weekly response time for each agent 

```python
performance.head()
```

```python
performance[['hh','mm','ss']]=performance["Average Response Time"].str.split(":",expand=True)
```

```python
performance['hh']=performance['hh'].astype(int)
performance['mm']=performance['mm'].astype(int)
performance['ss']=performance['ss'].astype(int)
```

```python
performance["Response time in second"]=(performance['hh']*3600) + (performance['mm']*60) + (performance['ss'])
performance.head()
```

```python
#9 . average weekly response time for each agent
AVG_res=performance.groupby(['Agent Name'])['Response time in second'].sum()/4.4
AVG_res.sort_values(ascending=False)
# avg_response time is in Seconds per week
```
